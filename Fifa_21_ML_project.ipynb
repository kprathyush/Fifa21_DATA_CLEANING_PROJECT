```python
import pandas as pd
fifa_data = pd.read_csv("fifa21_raw_data.csv")
print(fifa_data)
fifa_data.shape
```

```python
fifa_data.columns
```

```python
fifa_data.columns = fifa_data.columns.str.replace(' ', '_')
fifa_data.columns
```

```python
fifa_data.head(5)
```

```python
drop_col = ["photoUrl", "playerUrl"]
df = fifa_data.drop(columns=drop_col)
df.head(5)
```

```python
#dropping new line char
df["Team_&_Contract"] = df["Team_&_Contract"].str.replace("\n", " ")
df["Hits"] = df["Hits"].str.replace("\n", " ")
#df = df.applymap(lambda x: x.str.replace("\n", " "))
df.head(5)
```

```python
import numpy as np
from datetime import datetime
df1 = pd.DataFrame(df)
df1["Joined"] = pd.to_datetime(df1["Joined"], format = "%b %d, %Y")
curr_date = datetime.now()
for i, row in df.iterrows():
    dur = (curr_date - row['Joined']).days / 365.5
    if dur > 10:
        print("Player:", row['Name'], '|Playing since (in years): ', np.round(dur,1))
    
```

```python
value_regex = r"(\d+(\.\d*)?)([KkMm])"

# Remove the Euro symbol and extract numeric values with their suffixes
df['Value'] = df['Value'].str.replace("€", "")
df['Wage'] = df['Wage'].str.replace("€", "")
df['Release_Clause'] = df['Release_Clause'].str.replace("€", "")

value_extracted = df['Value'].str.extract(value_regex)
wage_extracted = df['Wage'].str.extract(value_regex)
release_clause_extracted = df['Release_Clause'].str.extract(value_regex)

# Convert the extracted values to numeric format and multiply by the appropriate multiplier
value_multiplier = {'K': 1000, 'k': 1000, 'M': 1000000, 'm': 1000000}
df['Value'] = value_extracted[0].astype(float) * value_extracted[2].str.lower().map(value_multiplier)
df['Wage'] = wage_extracted[0].astype(float) * wage_extracted[2].str.lower().map(value_multiplier)
df['Release_Clause'] = release_clause_extracted[0].astype(float) * release_clause_extracted[2].str.lower().map(value_multiplier)

# Print player information
for i, row in df.iterrows():
    print("Player:", row["Name"], "| Wage:", row["Wage"], "| Value:", row["Value"], "| Release_Clause:", row["Release_Clause"])
```

```python
#Remove ★ from below columns

replace_cols = ['W/F', 'SM', 'IR']

# Process columns by removing unwanted characters and converting to numeric
for col in replace_cols:
    if df[col].dtype == object:  # Check if the column is of object (string) type
        df[col] = df[col].str.replace('★', '').apply(pd.to_numeric)

print(df[replace_cols])
```

```python
import matplotlib.pyplot as plt
x = df['Wage']
y = df['Value']
plt.scatter(x,y)
```

```python
import matplotlib.pyplot as plt

# Filter players with low wages (you can adjust the threshold)
low_wage_threshold = df['Wage'].mean()  # Example threshold
low_wage_players = df[df['Wage'] < low_wage_threshold]

# Create a scatter plot
plt.figure(figsize=(10, 6))  
plt.scatter(low_wage_players['Wage'], low_wage_players['Value'], color='blue', alpha=0.5)
plt.xlabel('Wage')
plt.ylabel('Value')
plt.title('Scatter Plot: Wage vs Value for Low Wage Players')

# Annotate player names on the plot 
for i, player in low_wage_players.iterrows():
    plt.annotate(player['Name'], (player['Wage'], player['Value']), fontsize=8, alpha=0.7)

plt.show()
```
